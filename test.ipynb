```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")
#read csv
df = pd.read_csv('data.csv')
#print df content
df
#print df size
df.size
#print df columns
df.columns

#Preproccess of data


df.drop(df.loc[df[' TOTAL_NET_SALES '] == ' -   '].index, inplace=True)

df[' TOTAL_NET_SALES '] = df[' TOTAL_NET_SALES '].str.replace(',','').astype(float)


df.drop(df.loc[df[' sales_uplift_sku '] == ' -   '].index, inplace=True)

df[' sales_uplift_sku '] = df[' sales_uplift_sku '].str.replace(',','').astype(float)


df.drop(df.loc[df[' margin_uplift_sku '] == ' -   '].index, inplace=True)

df[' margin_uplift_sku '] = df[' margin_uplift_sku '].str.replace(',','').astype(float)


df.drop(df.loc[df[' sales_uplift_cg '] == ' -   '].index, inplace=True)

df[' sales_uplift_cg '] = df[' sales_uplift_cg '].str.replace(',','').astype(float)


df.drop(df.loc[df[' margin_uplift_cg '] == ' -   '].index, inplace=True)

df[' margin_uplift_cg '] = df[' margin_uplift_cg '].str.replace(',','').astype(float)


df.drop(df.loc[df[' sales_uplift_tce '] == ' -   '].index, inplace=True)

df[' sales_uplift_tce '] = df[' sales_uplift_tce '].str.replace(',','').astype(float)


df.drop(df.loc[df[' margin_uplift_tce '] == ' -   '].index, inplace=True)

df[' margin_uplift_tce '] = df[' margin_uplift_tce '].str.replace(',','').astype(float)


df['promo_start_date'] = pd.to_datetime(df['promo_start_date'])

df['promo_end_date'] = pd.to_datetime(df['promo_end_date'])




#print number of null values for columns of interest

print("Null values in Event New name column:"+str(df['Event new name'].isnull().values.sum()))

print("Null values in Promo Start Date column:"+str(df['promo_start_date'].isnull().values.sum()))

print("Null values in Promo End Date column:"+str(df['promo_end_date'].isnull().values.sum()))

print("Null values in JDA_2 column:"+str(df['JDA_2'].isnull().values.sum()))

print("Null values in BRAND NAME column:"+str(df['BRAND_NM'].isnull().values.sum()))

print("Null values in discount column:"+str(df['discount'].isnull().values.sum()))

print("Null values in Event type column:"+str(df['Event_type'].isnull().values.sum()))

print("Null values in total net sales column:"+str(df[' TOTAL_NET_SALES '].isnull().values.sum()))

print("Null values in sales_uplift_sku column:"+str(df[' sales_uplift_sku '].isnull().values.sum()))

print("Null values in margin_uplift_sku column:"+str(df[' margin_uplift_sku '].isnull().values.sum()))

print("Null values in sales_uplift_cg column:"+str(df[' sales_uplift_cg '].isnull().values.sum()))

print("Null values in  margin_uplift_cg  column:"+str(df[' margin_uplift_cg '].isnull().values.sum()))

print("Null values in  sales_uplift_tce column:"+str(df[' sales_uplift_tce '].isnull().values.sum()))

print("Null values in  margin_uplift_tce column:"+str(df[' margin_uplift_tce '].isnull().values.sum()))

print("Null values in  performarnce column:"+str(df['performance'].isnull().values.sum()))




plt.rcParams.update({'font.size': 6})

sns.countplot(x=df["Event new name"])

plt.show()


sns.countplot(x=df["JDA_2"])

plt.show()

df['BRAND_NM'].unique()


sns.countplot(y=df["BRAND_NM"])

plt.show()



sns.countplot(x=df["Event_type"])

plt.show()


sns.countplot(x=df["performance"])

plt.show()



df['discount'].describe()




sns.distplot(df['discount'],label='Discount Density of Distribution')

plt.legend()

plt.show()

df[' TOTAL_NET_SALES '].describe()

sns.distplot(df[' TOTAL_NET_SALES '],label='TOTAL NET SALES Density of Distribution')

plt.legend()

plt.show()

df[' sales_uplift_sku '].describe()
sns.distplot(df[' sales_uplift_sku '],label='Sales Uplift SKU Density of Distribution')

plt.legend()

plt.show()
df[' margin_uplift_sku '].describe()

sns.distplot(df[' margin_uplift_sku '],label='Margin Uplift SKU Density of Distribution')

plt.legend()

plt.show()

df[' sales_uplift_cg '].describe()
sns.distplot(df[' sales_uplift_cg '],label='Sales Uplift CG Density of Distribution')

plt.legend()

plt.show()
sns.distplot(df[' sales_uplift_cg '],label='Sales Uplift CG Density of Distribution')

plt.legend()

plt.show()
df[' margin_uplift_cg '].describe()

sns.distplot(df[' margin_uplift_cg '],label='Margin Uplift CG Density of Distribution')

plt.legend()

plt.show()


df[' sales_uplift_tce '].describe()
sns.distplot(df[' sales_uplift_tce '],label='Sales Uplift TCE Density of Distribution')

plt.legend()

plt.show()
df[' margin_uplift_tce '].describe()
sns.distplot(df[' margin_uplift_tce '],label='Margin Uplift TCE Density of Distribution')

plt.legend()

plt.show()

#calculate duration for earch promo 

e_date = df['promo_end_date'].to_numpy()

s_date = df['promo_start_date'].to_numpy()

duration = []


for i in range(len(e_date)):
    
    duration.append((e_date[i]-s_date[i])/(1000000000*86400))


sns.distplot(duration,label='Duration Density of Distribution')

plt.legend()

plt.show()
```
